## Plotting functions

```julia
using Plots
```

```julia
gr() # Use GR backend for plotting
unicodeplots() # To be used for quick visualization of data
```

```julia
plotly()  # Use Plotly backend for plotting
plot(dat[1:500,1], dat[1:500,2], label="line1")
scatter!(dat[1:500,1], dat[1:500,2], label="points1")
plot!(dat[1:500,1], dat[1:500,3], label="line2")
scatter!(dat[1:500,1], dat[1:500,3], label="points2")
xlabel!("Sepal Length")
ylabel!("Sepal Width")
title!("Sepal length vs Width")
```

### Creating multiple sub-plots

```julia
p1 = plot(dat[1:500,1],dat[1:500,2],theme(:dark)) # Make a line plot, change the theme of plots using Theme function
p2 = scatter(dat[1:500,1], dat[1:500,3]) # Make a scatter plot
p3 = plot(dat[1:500,1], dat[1:500,4], xlabel = "This one is labelled", lw = 3, title = "Subtitle")
p4 = histogram(dat[1:500,2], dat[1:500,6]) # Four histograms each with 10 points? Why not!


plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```
